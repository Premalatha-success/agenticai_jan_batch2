```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/refs/heads/main/CardioGoodFitness-1.csv")
```

```python
df.head()
```

```python
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"]=userdata.get('openai')
```

```python
from langchain_openai import OpenAI
```

```python
llm=ChatOpenAI(model="gpt-3.5-turbo",temperature=0)
```

```python
SYSTEM_PROMPT= """
You are an ADVANCED DATA ANALYST AGENT.
You must follow these rules strictly:

ROLE:

-Act as a Senior business data analyst.
-Answer only from the provided dataframe.

RULES:

- Never hallucinate.
-If data is missing: "Data not available in the dataset."
- Distinguish clearly between :
   .FACT (directly from data)
   .INSIGHT(logical inference)

REASONING RULES:

-When user asks Why?,give data backed reasoning.
- If there are multiple reasons,list them clearly.

OUTPUT RULES:

- Keep answers concise
- use bullets for insights
- Use simple language
- Avoid technical jorgan unless asked

SAFETY:
-Donot modify dataframe"""


```

```python
agent=create_pandas_dataframe_agent(llm=llm,df=df,verbose=True,
                                    agent_executor_kwargs={"system_message":SYSTEM_PROMPT},
                                    allow_dangerous_code=True)
```

```python
agent.run("How many rows are there?")
```

```python
agent.run("How many are using TM 798?")
```

```python
agent.run("What is the maximum age")
```
